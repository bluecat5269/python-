```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import webbrowser
from ipywidgets import interact
import sympy as sp
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC(gamma="auto")
```

```python
x=np.array([[6,6],[10,8],[3,7],[8,7],[9,3],[5,5],[6,9]])
plt.scatter(x[:,0],x[:,1])
y=np.array([1,2,1,2,1,1,2])
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=150);
```

```python
clf.fit(x,y)
```

```python
xr=np.random.rand(200)*5+4
yr=np.random.rand(200)*4+4
```

```python
plt.scatter(xr,yr)
```

```python
X=np.c_[xr,yr]
```

```python
Y=clf.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap="Paired")
```
